```python
import numpy as np
import matplotlib.pyplot as plt

# Set the time interval and the time step
T = 1
dt = 0.01

# Generate the Wiener process
t = np.arange(0, T + dt, dt)
n = len(t)
W = np.zeros(n)
for i in range(1, n):
    dW = np.random.normal(scale=np.sqrt(dt))
    W[i] = W[i-1] + dW

# Compute the increments and cumulative sum of the Wiener process
dW = np.diff(W)
W_sum = np.cumsum(dW)

# Compute the quadratic variation of the Wiener process
QV = T * np.ones(n-1)

# Plot the Wiener process, increments, and cumulative sum
fig, ax = plt.subplots(3, 1, figsize=(8, 10))
ax[0].plot(t, W)
ax[0].set_ylabel('Wiener process')
ax[1].plot(t[1:], dW)
ax[1].plot(t[1:], QV, 'r--')
ax[1].set_ylabel('Increments')
ax[2].plot(t[1:], W_sum)
ax[2].plot(t[1:], QV * t[1:], 'r--')
ax[2].set_ylabel('Cumulative sum')
ax[2].set_xlabel('Time')
plt.show()
```

```python
sum=0
for item in range(1,n):
    sum=sum+np.square(W[item]-W[item-1])
print(sum)
```

```python
import sympy as sym
c1=(float)(input("Enter Intital C: "))
r1=(float)(input("Enter Initial R: "))
n1=(float)(input("Enter Initial N: "))
c2=(float)(input("Enter Final C: "))
r2=(float)(input("Enter Final R: "))
n2=(float)(input("Enter Final N: "))
c,r,n=sym.symbols('c r n')
def A(c,r,n):
    return c*((1-(1+r)**(-n))/r)
f=A(c,r,n)
fc=f.diff(c)
fn=f.diff(n)
fr=f.diff(r)
f1=(fr.subs([(c,c1),(r,r1),(n,n1)]).evalf())*(r2-r1)
f2=(fc.subs([(c,c1),(r,r1),(n,n1)]).evalf())*(c2-c1)
f3=(fn.subs([(c,c1),(r,r1),(n,n1)]).evalf())*(n2-n1)
f1+f2+f3
```
